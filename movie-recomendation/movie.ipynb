```python
import pandas as pd
import numpy as np
```

```python
titles = pd.read_csv("./Movie_Id_Titles.csv")
rating = pd.read_csv("./Dataset.csv")
```

```python
df = pd.merge(titles, rating.drop("timestamp", axis=1))

df = df.sort_values("user_id")
```

```python
size = df["user_id"].nunique() * df["title"].nunique()

pre_pivot = {
    "user_id": [i for i in range(df["user_id"].nunique()) for j in range(df["title"].nunique())],
    "title": [i for j in range(df["user_id"].nunique()) for i in df["title"].unique()],
    "rating": [np.nan for i in range(size)]
}
```

```python
df.head()
print(df["user_id"].nunique())
```

```python
for i in range(df["user_id"].nunique()):
    id_title = ((list(df["title"].unique()).index(df["title"][i])+1) * ((df["user_id"][i]+1))) - 1
    pre_pivot["rating"][id_title] = df["rating"][i]
```

```python
new_df = pd.DataFrame(pre_pivot)
new_df.sort_values("rating", ascending=False).head(10)
```

```python
pivot = new_df.pivot(index="user_id", columns="title", values="rating")
pivot
```

```python
movie_watched = pivot["3 Ninjas: High Noon At Mega Mountain (1998)"]
similarity_with_other_movies = pivot.corrwith(movie_watched)
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
similarity_with_other_movies.head()
```
